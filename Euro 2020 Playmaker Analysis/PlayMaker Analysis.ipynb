## Characteristics of a Playmaker:
- **Great Passing Ability: Accuracy, Under Pressure etc**
- **Great Vision: Through Balls**
- **Good Dribbling Skills**
- **Assist Provider**
- **Key involvement in goals and shot creation**
- **Carrying the ball up the field**
- **Touches in Mid 3rd, attacking 3rd**

```python
dfChances = dfChances.groupby('Player').agg({'Player': 'first', 'team_name': 'first', 'Pos': 'first', 'Age': 'first', \
                                 'Min': 'sum', 'Ast': 'sum', 'SCA': 'sum', 'GCA': 'sum', 'PassCmp': 'sum', 'PassAtt': 'sum', \
                                'PassProg': 'sum'}).reset_index(drop = True)
```

**Calculating per90 stats**

```python
per90cols = ['Ast','SCA','GCA','PassCmp','PassAtt','PassProg']

for col in per90cols:
    dfChances[col + 'per90'] = dfChances[col].divide(dfChances['Min']).multiply(90)
dfChances.head(3)
```

## Metrics to consider for different abilities


**chances_creation_ability**: 'Astper90', 'SCAper90', 'GCAper90', 'KeyPassesper90', 'Opp1/3per90', 'PPAper90', 'CrsPAper90', 'PTTBper90', 'CPAper90'

**passing_ability**: 'PassCmpper90', 'PassAttper90', 'PassProgper90', 'PassAccuracy', 'PercentProgPasses', 'MedPassAcc', 'LongPassAcc', 'PTPressper90', 'PTSwper90'

**dribbling_ability**: 'DribSuccper90', 'DribAttper90', '#Playerslostper90', 'nutMegsper90'

**Possession_keeping**: 'TotTouchesper90', 'TouchMid 3rdper90', 'TouchAtt 3rdper90', 'TouchAtt Penper90', 'Carriesper90',
       'CarryTotDistper90', 'CarryPrgDistper90', 'CarryProgper90', 'Carry1/3per90'

### Chances Creation  Rating

**Chances_creation_rating**   =   'Astper90' * 0.28 + 'GCAper90' * 0.14 + 'SCAper90' * 0.105  + 'PTTBper90'* 0.07 + 'Opp1/3per90' * 0.035 + ('PPAper90' + 'CrsPAper90' + 'KeyPassesper90' + 'CPAper90') * 0.0875

```python
dffinal['Chances_creation_rating'] = dffinal['Astper90']*0.28 + dffinal['GCAper90']*0.14 + dffinal['SCAper90']*0.105 +\
                                    dffinal['PTTBper90']*0.07 + dffinal['Opp1/3per90']*0.035 + \
                                    (dffinal['PPAper90'] + dffinal['CrsPAper90'] +  \
                                     dffinal['KeyPassesper90'] + dffinal['CPAper90'])*0.0875

print('Top 10 Chance Creators in Euro2020 Group Stage:')
dffinal[['Player', 'team_name', 'Pos', 'Chances_creation_rating']].sort_values(by = 'Chances_creation_rating', ascending = False).head(10)
```

### Dribbling Ability Rating

**dribbling_ability_rating** = 'nutMegsper90' * 0.1 +  'DribAttper90' * 0.15 + 'DribSuccper90' * 0.25 + '#Playerslostper90' * 0.5 

```python
dffinal['dribbling_rating'] = dffinal['nutMegsper90']*0.1 + dffinal['DribAttper90']*0.15 + \
                                dffinal['DribSuccper90']*0.25 + dffinal['#Playerslostper90']*0.5

print('Top 10 Dribblers in Euro2020 Group Stage:')
dffinal[['Player','team_name','Pos','dribbling_rating']].sort_values(by = 'dribbling_rating', ascending = False).head(10)
```

### Passing Ability Rating

**Passing_ability_rating** = 'PassProgper90' * 0.16 + 'PercentProgPasses' * 0.16 + 'PassAccuracy' * 0.2 + 'MedPassAcc' * 0.08 + 'LongPassAcc' * 0.12 + 'PTPressper90' * 0.16 + 'PTSwper90' * 0.12

```python
dffinal['Passing_rating'] = dffinal['PassProgper90']*0.16 + dffinal['PercentProgPasses']*0.16 + dffinal['PassAccuracy']*0.2 + \
                dffinal['MedPassAcc']*0.08 + dffinal['LongPassAcc']*0.12 + dffinal['PTPressper90']*0.16 + \
                dffinal['PTSwper90']*0.12
# considering only those with more than 45 passes per90
print('Top 10 Passing Players in Euro2020 Group Stage:')
dffinal[dffinal['PassAttper90'] >= 45][['Player','team_name','Pos', 'Passing_rating']].sort_values(by = 'Passing_rating', ascending = False).head(10)
```

### Possession Keepers

**Possession_rating** = 'TouchMid 3rdper90' * 0.04 + 'TouchAtt 3rdper90' * 0.08 + 'TouchAtt Penper90' * 0.12 + 'Carriesper90' * 0.12 + 'CarryTotDistper90' * 0.12 + 'CarryPrgDistper90' * 0.16 + 'CarryProgper90' * 0.16 + 'Carry1/3per90 * 0.2

```python
dffinal['Possession_keeping'] = dffinal['TouchMid 3rdper90']*0.04 + dffinal['TouchAtt 3rdper90']*0.08 + \
                                dffinal['TouchAtt Penper90']*0.12 + dffinal['Carriesper90']*0.12 + \
                                dffinal['CarryTotDistper90']*0.12 + dffinal['CarryPrgDistper90']*0.16 + \
                                dffinal['CarryProgper90']*0.16 + dffinal['Carry1/3per90']*0.2
# graeter than 60 touches per90
print('Top 10 Possession keepers of Euro 2020 Group Stage:')
dffinal[dffinal['TotTouchesper90'] >= 60][['Player','team_name','Pos', 'Possession_keeping']].sort_values(by = 'Possession_keeping', ascending = False).head(10)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler((1,10))
data = dffinal.copy()
data[['Chances_creation_rating','dribbling_rating','Passing_rating','Possession_keeping']] = scaler.fit_transform(data[['Chances_creation_rating','dribbling_rating','Passing_rating','Possession_keeping']])
data.head()
```

```python
rounding = [ 'Astper90','SCAper90','GCAper90','PassCmpper90','PassAttper90','PassProgper90','PassAccuracy', \
            'PercentProgPasses','KeyPassesper90','Opp1/3per90','PPAper90','CrsPAper90','MedPassAcc','LongPassAcc', \
            'PTTBper90','PTPressper90','PTSwper90','TotTouchesper90','TouchMid 3rdper90','TouchAtt 3rdper90', \
            'TouchAtt Penper90','DribSuccper90','DribAttper90','#Playerslostper90','nutMegsper90','Carriesper90', \
            'CarryTotDistper90','CarryPrgDistper90','CarryProgper90','Carry1/3per90','CPAper90']

r = data.copy()
data[rounding] = np.round(data[rounding],2)

data.head()
```

```python
data['Age'] = data['Age'].apply(lambda x: x.split('-')[0])
data.head()
```

## PlayMaker Rating

**Playmaker_rating** = Chances_creation_rating * 0.4 + dribbling_rating * 0.1 + Passing_rating * 0.25 + Possession_keeping * 0.25

```python
data['PlayMaker Rating'] = data['Chances_creation_rating']*0.4 + data['dribbling_rating']*0.1 + \
                            data['Passing_rating']*0.25 + data['Possession_keeping']*0.25

data[data['Min'] >= 45][['Player','team_name','Pos', 'PlayMaker Rating']].sort_values(by = 'PlayMaker Rating', ascending = False).head(5)
```

```python
data['PlayMaker Rating'] = scaler.fit_transform(data[['PlayMaker Rating']])
```

```python
print('Top 10 Playmakers of Euro 2020:')

data[['Player','team_name','Pos', 'PlayMaker Rating']].sort_values(by = 'PlayMaker Rating', ascending = False).head(10)
```
